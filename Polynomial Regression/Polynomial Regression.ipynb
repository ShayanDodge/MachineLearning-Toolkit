# The Python code for Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import operator

np.random.seed(0)

# Generate a new dataset
x = 2 * np.random.normal(0, 1, 50)
y = 2 * x + 3 * (x ** 2) + np.random.normal(0, 3, 50)

# Data is transformed to include another axis
x = x[:, np.newaxis]

polyfeatures = PolynomialFeatures(degree=2)
xpoly = polyfeatures.fit_transform(x)
print(xpoly)

model = LinearRegression()
model.fit(xpoly, y)
ypoly_pred = model.predict(xpoly)

rmse = np.sqrt(mean_squared_error(y, ypoly_pred))
r2 = r2_score(y, ypoly_pred)

print("RMSE:", rmse)
print("R2 Score:", r2)

plt.scatter(x, y, s=20)

# The values of x are sorted according to the degree before the line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x[:, 0], ypoly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)

plt.plot(x, y_poly_pred, color='m')
plt.show()
```
